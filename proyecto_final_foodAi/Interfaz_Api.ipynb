Voy a crear el codigo donde preparo la interfaz y la guardo en un archivo llamado app.py. Esta celda guarda el c√≥digo de una interfaz en Streamlit para la predicci√≥n de platos basada en im√°genes. Permite subir una imagen de un plato, la env√≠a a una API para su an√°lisis, luego muestra el nombre del plato, sus calor√≠as, macronutrientes, ingredientes y un enlace a la receta.  
Finalmente, el c√≥digo se guarda en la ruta especificada para su posterior ejecuci√≥n.  

**Es importante que se est√© ejecutando el c√≥digo donde tenemos la API, porque sino esta corriendo no funcionar√° y adem√°s tiene que coincidir la API_URL que te arroja ngrok con la que se pone en este c√≥digo y siempre que corras de nuevo la API, poner la URL ue te arroja en ese momento**


```python
# Ruta donde se guardar√° el archivo
ruta_guardado = "/content/drive/MyDrive/tokio/deep_learning/proyecto_final_foodAi/app.py"

# C√≥digo de la interfaz Streamlit
codigo = """import streamlit as st
import requests
import json
from PIL import Image

# URL de la API
API_URL = "https://fe4a-104-196-143-67.ngrok-free.app"

st.set_page_config(page_title="Predicci√≥n de Platos", page_icon="üçΩÔ∏è", layout="wide")

st.title("üçΩÔ∏è Predicci√≥n de Platos")

uploaded_file = st.file_uploader("üì∑ Sube una imagen del plato", type=["jpg", "png", "jpeg"])

if uploaded_file:
    image = Image.open(uploaded_file)
    st.image(image, caption="Imagen subida", use_container_width=True)

    with st.spinner("‚è≥ Analizando imagen..."):
        files = {"file": uploaded_file.getvalue()}
        response = requests.post(f"{API_URL}/predict", files=files)

    if response.status_code == 200:
        result = response.json()
        st.success("‚úÖ Predicci√≥n realizada con √©xito")

        st.markdown("### üçΩÔ∏è Plato:")
        st.markdown(f"<p style='font-size: 22px; font-weight: bold;'>{result['plato']}</p>", unsafe_allow_html=True)

        st.markdown("### üî• Calor√≠as:")
        st.markdown(f"<p style='font-size: 22px; font-weight: bold;'>{result['calorias']} kcal</p>", unsafe_allow_html=True)

        st.markdown("### üõ¢Ô∏è Grasas:")
        st.markdown(f"<p style='font-size: 22px; font-weight: bold;'>{result['grasas']} g</p>", unsafe_allow_html=True)

        st.markdown("### üçû Carbohidratos:")
        st.markdown(f"<p style='font-size: 22px; font-weight: bold;'>{result['carbohidratos']} g</p>", unsafe_allow_html=True)

        st.markdown("### üí™ Prote√≠nas:")
        st.markdown(f"<p style='font-size: 22px; font-weight: bold;'>{result['proteinas']} g</p>", unsafe_allow_html=True)

        # Mostrar ingredientes correctamente alineados
        if result.get("ingredientes") and result["ingredientes"] != "No disponible":
            st.markdown("### üìú Ingredientes:")
            ingredientes_lista = result["ingredientes"].splitlines()
            ingredientes_limpios = [ing.strip("- ").strip() for ing in ingredientes_lista if ing.strip()]
            for ing in ingredientes_limpios:
                st.markdown(f"- {ing}")

        # Mostrar enlace de la receta si est√° disponible
        if result.get("url_receta") and result["url_receta"] != "No disponible":
            st.markdown(f"üìå **[Ver Receta]({result['url_receta']})**", unsafe_allow_html=True)

    else:
        st.error("‚ùå Error en la predicci√≥n")
"""

# Guardo el c√≥digo en un archivo
with open(ruta_guardado, "w", encoding="utf-8") as f:
    f.write(codigo)

print(f"‚úÖ Archivo guardado en {ruta_guardado}")
```
